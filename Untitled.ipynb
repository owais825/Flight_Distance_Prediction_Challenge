```python
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in km
    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance
```

```python
train_df['Haversine Distance'] = haversine_distance(train_df['Origin Latitude'], train_df['Origin Longitude'], train_df['Destination Latitude'], train_df['Destination Longitude'])
test_df['Haversine Distance'] = haversine_distance(test_df['Origin Latitude'], test_df['Origin Longitude'], test_df['Destination Latitude'], test_df['Destination Longitude'])

train_df['Distance Difference'] = abs(train_df['Great Circle Distance'] - train_df['Haversine Distance'])
test_df['Distance Difference'] = abs(test_df['Great Circle Distance'] - test_df['Haversine Distance'])
```
